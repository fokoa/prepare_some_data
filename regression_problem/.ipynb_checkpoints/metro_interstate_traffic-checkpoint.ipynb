## 3. Processing

### 3.1 temp

```python
# Let's convert temperature to kelvin into celsius
data['temp'] = data['temp'].apply(lambda x : x-273.15 if x else x)
```

```python
data.temp.isna().sum()
```

### 3.2. rain_1h

```python
# Descriptive statistics
data[['rain_1h']].describe().T
```

```python
data['rain_1h'].replace(9831.30, data['rain_1h'].median(), inplace=True)
```

```python
# Let's bin rain_1h
bins = [-1, 0.1, 0.26, 0.51, 1, 100000]
labels = [1, 2, 3, 4, 5];
data['rain_bin'] = pd.cut(data['rain_1h'], bins=bins, labels=labels, right=False);
```

```python
data[(data.rain_1h>=0.51) & (data.rain_1h<=1.0)].shape
```

### 3.3. snow_1h

```python
# Descriptive statistics
data[['snow_1h']].describe().T
```

```python
data[['snow_1h']].value_counts()
```

```python
# Let's bin snow_1h
bins = [-1, 0.1, 0.52];
labels = [1, 2];
data['snow_bin'] = pd.cut(data['snow_1h'], bins=bins, labels=labels, right=False);
```

### 3.4. clouds_all

```python
data['clouds_all'].describe()
```

```python
# Let's bin clouds_all
bins = [-1, 26, 51, 76, 101];
labels = [1, 2, 3, 4];
data['clouds_bin'] = pd.cut(data['clouds_all'], bins=bins, labels=labels, right=False);
```

### 3.5. date_time

```python
# Extract date features from the date columns
data['date_time'] = pd.to_datetime(data['date_time'])
for date_feature in ['year', 'month', 'day']:
    data[date_feature] = getattr(data['date_time'].dt, date_feature)
```

```python
data.head()
```

## 4. Transformation

```python
# Categorical
data = pd.get_dummies(data, columns=['weather_main', 'weather_description', 'holiday']); # Nominal
data['year'] = LabelEncoder().fit_transform(data['year']); # Ordinal
```

```python
# Numerical
data[['temp', 'rain_1h', 'snow_1h', 'clouds_all']] = StandardScaler().fit_transform(data[['temp', 'rain_1h', 'snow_1h', 'clouds_all']])
data.head()
```

## 5. Split data & Fitting models

```python
# Select main columns to be used in training
main_cols = data.columns.difference(['date_time', 'hour', 'traffic_volume']); 
X = data[main_cols];
y = data['traffic_volume']
```

```python
# Split out train and test dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42);

# Cross validation
kfold = KFold(n_splits=10,  shuffle=True, random_state=21);
```

```python
# Fitting
linears = [];
linears.append(('LR', LinearRegression()));
linears.append(('RIDGE', Ridge()));
linears.append(('LASSO', Lasso()));

# Evaluate
for name, model in linears:
    scores = cross_val_score(model, X_train, y_train, scoring='neg_root_mean_squared_error', cv=kfold);
    print('%s : %.4f(%.4f)' % (name, -1 * scores.mean(), scores.std()));
```

## 6. Evaluate on test data

```python
# Fit the best model
mod = Lasso();
mod = mod.fit(X_train, y_train);
y_pred = mod.predict(X_test);
```

```python
# Evaluate each model in turn with kfold
print('Score :', mean_squared_error(y_test, y_pred, squared=False));
```
